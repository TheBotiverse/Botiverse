```python
from botiverse.bots import SpeechClassifier
from botiverse.bots.Vocalizer.utils import voice_input
```

```python
S = SpeechClassifier(['Yes', 'No'], samplerate=16000, duration=0.9, repr='wav2vec')
X, y = S.generate_read_data(force_download_noise=False, n=5)
```

```python
S.fit(X, y, hidden=128, patience=100, α=0.01)
#S.save('speechclassifier')
```

```python
#S.load('speechclassifier')
from tqdm.notebook import tqdm
yes_preds = []
# for each .wav in test/yes, predict the class
import os
i = 0
for file in tqdm(os.listdir('test/yes')):
    if file.endswith('.wav'):
        pred, prob = S.predict('test/yes/'+file)
        yes_preds.append(pred)
        i += 1
        
    if i == 100: break

# compute the recall

no_preds = []
# for each .wav in test/no, predict the class
import os
i = 0
for file in tqdm(os.listdir('test/no')):
    if file.endswith('.wav'):
        pred, prob = S.predict('test/no/'+file)
        no_preds.append(pred)
        i+=1
        
    if i == 100: break
        
print(yes_preds)
print(no_preds)
```

```python
# compute the recall of the yes class by counting strings with 'Yes' in them
yes_recall = sum([1 for pred in yes_preds if 'Yes' in pred])/len(yes_preds)
# compute the recall of the no class by counting strings with 'No' in them
no_recall = sum([1 for pred in no_preds if 'No' in pred])/len(no_preds)

print(f"Recall of the yes class: {yes_recall} and recall of the no class: {no_recall}")
```
